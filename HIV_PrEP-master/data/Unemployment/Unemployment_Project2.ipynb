```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Loading csv files
file = "unemployment _2008_2018.csv"
```

```python
# Read CSV files
unemployment = pd.read_csv(file)
unemployment.head()
```

```python
# Creating a filtered dataframe from specific columns
unemployment= unemployment[['Area_name', "Unemployment_rate_2008", "Unemployment_rate_2009" , "Unemployment_rate_2010", "Unemployment_rate_2011", "Unemployment_rate_2012", "Unemployment_rate_2013", "Unemployment_rate_2014" , "Unemployment_rate_2015" ,"Unemployment_rate_2016" , "Unemployment_rate_2017", "Unemployment_rate_2018"]]
```

```python
States=["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "District of Columbia", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]
State_unemployment= unemployment[unemployment.Area_name.isin(States)]
State_unemployment= State_unemployment.rename(columns={'Area_name': 'Name'})
State_unemployment.drop_duplicates(subset= "Name", keep = 'first', inplace = True)
State_unemployment.head(50)
```

```python
State_unemployment['Unemployment_rate_2008-2012'] = State_unemployment[['Unemployment_rate_2008','Unemployment_rate_2009', 'Unemployment_rate_2010', 'Unemployment_rate_2011','Unemployment_rate_2012']].mean(axis=1)
State_unemployment['Unemployment_rate_2013-2018'] = State_unemployment[['Unemployment_rate_2013','Unemployment_rate_2014', 'Unemployment_rate_2015', 'Unemployment_rate_2016','Unemployment_rate_2017','Unemployment_rate_2018']].mean(axis=1)
State_unemployment.head()
```

```python
State_unemployment= State_unemployment.drop(['Unemployment_rate_2008','Unemployment_rate_2009', 'Unemployment_rate_2010', 'Unemployment_rate_2011','Unemployment_rate_2012','Unemployment_rate_2013','Unemployment_rate_2014','Unemployment_rate_2015', 'Unemployment_rate_2016','Unemployment_rate_2017','Unemployment_rate_2018'], axis=1)
State_unemployment.head()
```

```python
State_unemployment["Unemployment_rate_variance"] = State_unemployment["Unemployment_rate_2008-2012"]-State_unemployment["Unemployment_rate_2013-2018"]
```

```python
State_unemployment.head()
```

```python
file = "New_HIV_Case_Rates.csv"
cases = pd.read_csv(file)
cases.head()
```

```python
State_unemployment = State_unemployment.merge(cases, on='Name')
State_unemployment.head()
```

```python
State_unemployment.to_csv('unemployment_rate.csv', index = False, na_rep= '')
```
