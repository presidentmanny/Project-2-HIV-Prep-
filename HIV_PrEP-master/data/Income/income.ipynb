```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Loading csv files
file = "Income _2008_2018.csv"
```

```python
# Read CSV files
income = pd.read_csv(file)
income.head(1000)
```

```python
income['Annual_Avg_Pay_2008-2012'] = income[['Annual_Avg_Pay_2008','Annual_Avg_Pay_2009', 'Annual_Avg_Pay_2010', 'Annual_Avg_Pay_2011','Annual_Avg_Pay_2012']].mean(axis=1)
income['Annual_Avg_Pay_2013-2018'] = income[['Annual_Avg_Pay_2013','Annual_Avg_Pay_2014', 'Annual_Avg_Pay_2015', 'Annual_Avg_Pay_2016','Annual_Avg_Pay_2017','Annual_Avg_Pay_2018']].mean(axis=1)
income.head()
```

```python
income= income.drop(['Annual_Avg_Pay_2008','Annual_Avg_Pay_2009', 'Annual_Avg_Pay_2010', 'Annual_Avg_Pay_2011','Annual_Avg_Pay_2012','Annual_Avg_Pay_2013','Annual_Avg_Pay_2014','Annual_Avg_Pay_2015', 'Annual_Avg_Pay_2016','Annual_Avg_Pay_2017','Annual_Avg_Pay_2018'], axis=1)
income.head()
```

```python
income["Annual_Avg_Pay_Variance"] = income["Annual_Avg_Pay_2013-2018"]-income["Annual_Avg_Pay_2008-2012"]
income.head()
```

```python
income["Name"] = income["Name"].str.strip()
income.head()
```

```python
file = "New_HIV_Case_Rates.csv"
cases = pd.read_csv(file)
cases.head(20)
```

```python
income = income.merge(cases, on='Name')
```

```python
income.to_csv('Annual_Avg_Pay.csv', index = False, na_rep= '')
```
