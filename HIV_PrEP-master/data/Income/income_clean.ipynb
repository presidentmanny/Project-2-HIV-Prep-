```python
#Import needed libraries
import pandas as pd
import glob 
```

```python
#Define locations of datafiles using glob 
income_files = glob.glob('resources/income/*.xlsx')
print(income_files)
```

```python
#Read in the files by looping through files and then appending to  a list
income_dfs = [] 

for file in income_files:
    income_dfs.append(pd.read_excel(file))
```

```python
#Concatanate income_dfs to one df
concat_income_df = pd.concat(income_dfs, ignore_index=True)

concat_income_df.head()
```

```python
concat_income_df.columns
```

```python
total_income = concat_income_df[["Area\nCode","Area Type","Year","Area","Ownership",
                                  "Annual Average Weekly Wage","Annual Average Pay"]]

total_income.head()
```

```python
tot_income = total_income.rename(columns = {"Area\nCode":"Area Code", 
                               "Annual Average Weekly Wage":"Annual Avg Weekly Wage", 
                               "Annual Average Pay":"Annual Avg Pay"})
tot_income.head()
```

```python
income_states = tot_income[(tot_income["Ownership"]=="Total Covered") & 
                            (tot_income["Area Type"].str.contains("Nation|State"))]
income_states.head()
```

```python
#Reset index
income_states= income_states.reset_index(drop=True)
income_states.head()
```

```python
income_states["Area"] = income_states["Area"].str.replace("-- Statewide", " ")
income_states.head()
```

```python
income_by_state = income_states[["Area","Year","Annual Avg Pay","Annual Avg Weekly Wage","Area Type","Area Code"]]
income_by_state.head()
```

```python
income_by_state.to_csv('resources/income_by_state.csv',index=False)
```
